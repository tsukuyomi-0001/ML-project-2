# Pipeline

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, FunctionTransformer
from sklearn.compose import make_column_selector, make_column_transformer
```

```python
obj_pipeline = make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    OneHotEncoder(handle_unknown='ignore')
)

num_pipeline = make_pipeline(
    SimpleImputer(strategy='most_frequent')
)

def datetime_features_out(data):
    df = data.copy()
    df.columns = ['Date']

    df['year'] = pd.to_datetime(df['Date']).dt.year
    df['month'] = pd.to_datetime(df['Date']).dt.month
    df['day'] = pd.to_datetime(df['Date']).dt.day

    return df.drop(columns='Date')
 
datetime_pipeline = make_pipeline(
    FunctionTransformer(datetime_features_out, validate=False)
)

pipeline = make_column_transformer(
    (obj_pipeline, make_column_selector(dtype_include=object)),
    (SimpleImputer(strategy='most_frequent'), make_column_selector(dtype_include=int)),
    (datetime_pipeline, ['Date'])
)

def log_transform(data):
    df = data.copy()
    epsilon = 1e-8
    df = np.log1p(df + epsilon)

    return df 

label_pipeline = make_pipeline(
    FunctionTransformer(log_transform, validate=False)
)
```

# Train, test, valid

```python
new_df.head()
```

```python
new_df[new_df['Country/Region'] == 'India']
```

```python
train_set = pd.DataFrame()
test_set = pd.DataFrame()
valid_set = pd.DataFrame()

for country in new_df['Country/Region'].unique():
    full_data = new_df[new_df['Country/Region'] == country]

    n = len(full_data)
    train_end = int(n * 0.85)
    valid_end = int(n * 0.95)

    train_data = full_data.iloc[:train_end]
    valid_data = full_data.iloc[train_end:valid_end]
    test_data  = full_data.iloc[valid_end:]

    train_set = pd.concat([train_set, train_data], ignore_index=True)
    valid_set = pd.concat([valid_set, valid_data], ignore_index=True)
    test_set  = pd.concat([test_set, test_data], ignore_index=True)
```

```python
X_train = pipeline.fit_transform(train_set)
y_train = label_pipeline.fit_transform(train_set['Deaths']) 
# y_train = train_set['Deaths']

X_valid = pipeline.transform(valid_set)
y_valid = label_pipeline.transform(valid_set['Deaths']) 
# y_valid = valid_set['Deaths']  

x_test = pipeline.transform(test_set)
y_test =  label_pipeline.transform(test_set['Deaths']) 
# y_test = test_set['Deaths']  
```

# Model

```python
import xgboost as xgb 

model = xgb.XGBRegressor(eval_metric='rmse')
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import root_mean_squared_error


# no log transform to death: 2352.19873046875 
# with np.log1p transform to death: 4709.838953731951
pred_log = model.predict(X_valid)
pred = np.expm1(pred_log) 
rmse_for_valid = root_mean_squared_error(np.expm1(y_valid), pred) 

# pred = model.predict(X_valid)
# rmse_for_valid = root_mean_squared_error(ny_valid, pred)
rmse_for_valid
```
